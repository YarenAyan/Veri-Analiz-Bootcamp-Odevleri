### Problem Definition and Objectives
This project aims to analyze UK traffic accident data between 2000 and 2016. The dataset includes over 1.6 million accidents based on police reports and traffic flow data. The main objectives of this analysis are:
- To examine how changes in traffic flow impact accidents
- To identify factors that increase accident rates
- To build models that can predict accident rates over time
- To explore differences in accident patterns between urban and rural areas
- To understand regional differences (England, Scotland, Wales)


## 1. Importing Libraries & Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load datasets
traffic_df = pd.read_csv('ukTrafficAADF.csv')
accidents_1 = pd.read_csv('accidents_2005_to_2007.csv')
accidents_2 = pd.read_csv('accidents_2009_to_2011.csv')
accidents_3 = pd.read_csv('accidents_2012_to_2014.csv')

# Merge accident data
accidents_df = pd.concat([accidents_1, accidents_2, accidents_3], ignore_index=True)
```

## 2. Data Overview & Preprocessing

```python
# View basic info
print(accidents_df.info())
print(traffic_df.info())

# Convert date to datetime
accidents_df['Date'] = pd.to_datetime(accidents_df['Date'], errors='coerce')
accidents_df['Year'] = accidents_df['Date'].dt.year
```

## 3. Exploratory Data Analysis

```python
# Yearly accident trends
yearly_accidents = accidents_df['Year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
sns.lineplot(x=yearly_accidents.index, y=yearly_accidents.values)
plt.title('Yearly Accident Counts')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

## 5. Urban vs Rural & Region Analysis

```python
# Analyze by Road_Type and Region
urban_rural = accidents_df['Road_Type'].value_counts()
sns.barplot(x=urban_rural.index, y=urban_rural.values)
plt.title('Urban vs Rural Road Accidents')
plt.xticks(rotation=45)
plt.show()
```

## 6. Predictive Modeling

```python
# Example: Predict number of accidents by year
from sklearn.linear_model import LinearRegression

X = yearly_accidents.index.values.reshape(-1, 1)
y = yearly_accidents.values
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

plt.plot(X, y, label='Actual')
plt.plot(X, predictions, label='Predicted', linestyle='--')
plt.title('Linear Regression on Yearly Accidents')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.legend()
plt.show()
```
